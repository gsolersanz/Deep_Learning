## **First Implementation:**

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class IntermediateCNN(nn.Module):
    def __init__(self, num_classes=3):
        super(IntermediateCNN, self).__init__()

        # Bloque 1: de 3 canales a 64
        self.block1 = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        # Bloque 2: de 64 a 128
        self.block2 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        # Bloque 3: de 128 a 256
        self.block3 = nn.Sequential(
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        # Bloque 4: de 256 a 512
        self.block4 = nn.Sequential(
            nn.Conv2d(256, 512, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        # Para un input de 224x224, luego de 4 pooling (división por 16) obtenemos dimensiones 14x14.
        self.classifier = nn.Sequential(
            nn.Linear(512 * 14 * 14, 4096),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(4096, 4096),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(4096, num_classes)
        )

    def forward(self, x):
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        x = self.block4(x)
        # Aplanamos las características para pasarlas a la red fully connected
        x = x.view(x.size(0), -1)
        x = self.classifier(x)
        return x

# Crear instancia del modelo y mostrar resumen
model = IntermediateCNN(num_classes=3)
print(model)
```

## **Second Implementation:**

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ResIncepNet(nn.Module):
    def __init__(self, num_classes=3):
        super(ResIncepNet, self).__init__()
        # Stem: large convolution + batch norm + ReLU + max pooling (ResNet style)
        self.conv1 = nn.Conv2d(3, 64, kernel_size=7, stride=2, padding=3)
        self.bn1   = nn.BatchNorm2d(64)
        self.pool1 = nn.MaxPool2d(kernel_size=3, stride=2, padding=1)

        # Inception-style multi-branch block (all operations are done in forward)
        # We'll split channels into 4 branches (each outputs 16 channels to total 64)
        # Branch 1: 1x1 conv
        self.branch1_conv = nn.Conv2d(64, 16, kernel_size=1)

        # Branch 2: 1x1 conv then 3x3 conv
        self.branch2_conv1 = nn.Conv2d(64, 16, kernel_size=1)
        self.branch2_conv2 = nn.Conv2d(16, 16, kernel_size=3, padding=1)

        # Branch 3: 1x1 conv then two 3x3 convs
        self.branch3_conv1 = nn.Conv2d(64, 16, kernel_size=1)
        self.branch3_conv2 = nn.Conv2d(16, 16, kernel_size=3, padding=1)
        self.branch3_conv3 = nn.Conv2d(16, 16, kernel_size=3, padding=1)

        # Branch 4: 3x3 max pool then 1x1 conv
        self.branch4_pool  = nn.MaxPool2d(kernel_size=3, stride=1, padding=1)
        self.branch4_conv  = nn.Conv2d(64, 16, kernel_size=1)

        # Linear projection: after concatenation (16*4 = 64 channels) project back to 64 channels
        self.conv_linear = nn.Conv2d(64, 64, kernel_size=1)

        # Classifier: global average pooling then a fully-connected layer
        self.fc = nn.Linear(64, num_classes)

    def forward(self, x):
        # Stem block
        x = self.conv1(x)
        x = self.bn1(x)
        x = F.relu(x)
        x = self.pool1(x)  # output: [B, 64, H/4, W/4]

        # Inception-style branch with residual connection (all in one forward function)
        # Branch 1
        b1 = self.branch1_conv(x)
        # Branch 2
        b2 = self.branch2_conv1(x)
        b2 = F.relu(b2)
        b2 = self.branch2_conv2(b2)
        # Branch 3
        b3 = self.branch3_conv1(x)
        b3 = F.relu(b3)
        b3 = self.branch3_conv2(b3)
        b3 = F.relu(b3)
        b3 = self.branch3_conv3(b3)
        # Branch 4
        b4 = self.branch4_pool(x)
        b4 = self.branch4_conv(b4)

        # Concatenate branches along channel dimension
        out = torch.cat([b1, b2, b3, b4], dim=1)
        # Project back to 64 channels and add residual connection
        out = self.conv_linear(out)
        out = out + x
        out = F.relu(out)

        # Global average pooling and classification
        out = F.adaptive_avg_pool2d(out, (1, 1))
        out = out.view(out.size(0), -1)
        out = self.fc(out)
        return out

# Create and print the network
model = ResIncepNet(num_classes=3)
print(model)
```

# Dataset and Training

Now we are going to train one of the implemented networks. For this, we will use the CIFAR-10 dataset. This dataset is composed of 10 classes but we will only use 3.

```python
#@title Imports
import glob
import random
import numpy as np
import pandas as pd

import torch
from torch import nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset

torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
#@title Dataset

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)) # Normalize to [-1, 1] range
])

def filter_indices_with_limits(dataset, class_sample_limits):
    indices = []
    class_counts = {cls: 0 for cls in class_sample_limits.keys()}
    for i in range(len(dataset)):
        _, label = dataset[i]
        class_name = dataset.classes[label]
        if class_name in class_sample_limits:
            if class_counts[class_name] < class_sample_limits[class_name]:
                indices.append(i)
                class_counts[class_name] += 1
            if all(class_counts[cls] >= limit for cls, limit in class_sample_limits.items()):
                break
    return indices


trainset = datasets.CIFAR10(root='CIFAR10_data/', train=True, download=True, transform=transform)
testset = datasets.CIFAR10(root='CIFAR10_data/', train=False, download=True, transform=transform)

train_limits = {'airplane': 500, 'automobile': 500, 'bird': 50}
test_limits = {'airplane': 100, 'automobile': 100, 'bird': 100}

train_indices = filter_indices_with_limits(trainset, train_limits)
trainset_filtered = Subset(trainset, train_indices)

test_indices = filter_indices_with_limits(testset, test_limits)
testset_filtered = Subset(testset, test_indices)

trainloader = DataLoader(trainset_filtered, batch_size=64, shuffle=True)
testloader = DataLoader(testset_filtered, batch_size=64, shuffle=True)
```

## Data Augmentation

As you can see, we have a dataset that is not balanced. The `bird` class only has 50 samples, while `airplane` and `automobile` have 500. In this section, you will need to implement a data augmentation technique to solve this problem in our dataset.

You can find examples of different techniques at the following link:

1. [Getting started with transforms v2](https://pytorch.org/vision/stable/auto_examples/transforms/plot_transforms_getting_started.html#sphx-glr-auto-examples-transforms-plot-transforms-getting-started-py)
2. [Transforming and augmenting images](https://pytorch.org/vision/stable/transforms.html)


### **Apply Data Augmentation**

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import Dataset, DataLoader, Subset

# Definir la transformación base y la de augmentación para la clase minoritaria
base_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

augmentation_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),  # rotación aleatoria de ±15 grados
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Cargar el dataset CIFAR10 sin transformaciones (para aplicar luego las nuestras)
trainset_full = datasets.CIFAR10(root='CIFAR10_data/', train=True, download=True, transform=None)

# Filtrar índices según límites (500 para 'airplane' y 'automobile', 50 para 'bird')
train_limits = {'airplane': 500, 'automobile': 500, 'bird': 50}
train_indices = filter_indices_with_limits(trainset_full, train_limits)
trainset_filtered = Subset(trainset_full, train_indices)

# Dataset personalizado que aplica data augmentation a la clase minoritaria ("bird")
class BalancedAugmentedDataset(Dataset):
    def __init__(self, subset_dataset, base_transform, augmentation_transform, minority_class="bird", augmentation_factor=10):
        self.subset_dataset = subset_dataset
        self.base_transform = base_transform
        self.augmentation_transform = augmentation_transform
        self.minority_class = minority_class
        self.augmentation_factor = augmentation_factor

        # Generamos una lista de tuplas: (índice en subset_dataset, flag para saber si es de minority)
        self.indices = []
        for idx in range(len(self.subset_dataset)):
            image, label = self.subset_dataset[idx]
            # Obtenemos el nombre de la clase usando el atributo "classes" del dataset original
            class_name = self.subset_dataset.dataset.classes[label]
            if class_name == self.minority_class:
                # Replicamos la muestra augmentation_factor veces
                self.indices.extend([(idx, True)] * self.augmentation_factor)
            else:
                self.indices.append((idx, False))

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, index):
        idx, is_minority = self.indices[index]
        image, label = self.subset_dataset[idx]
        # Aplicamos la transformación correspondiente
        if is_minority:
            image = self.augmentation_transform(image)
        else:
            image = self.base_transform(image)
        return image, label

# Crear el dataset balanceado con data augmentation
balanced_trainset = BalancedAugmentedDataset(trainset_filtered, base_transform, augmentation_transform,
                                             minority_class="bird", augmentation_factor=10)

# Crear el DataLoader
trainloader = DataLoader(balanced_trainset, batch_size=64, shuffle=True)
```

# Fine-Tunning

Now we are going to perform fine-tuning on the two architectures that have been implemented.

Although the network architecture is a blend of AlexNet and VGG19, I chose to rely exclusively on VGG19’s pretrained weights for the fine-tuning step. There is no explicit requirement to incorporate both pretrained models. Therefore, to ensure a coherent transfer learning approach, I opted for VGG19 alone. While I considered including AlexNet as well, the decision ultimately rested on the practical limitation of development time rather than any theoretical preference for one model over the other.

```python
#Fine-Tunning de IntermediateCNN - vgg19
# Because of the architecture that we have we choose to do fine-tuning only with the pretrained model vgg19

model = IntermediateCNN(num_classes=3)

vgg19_pretrained = models.vgg19(pretrained=True)
vgg19_features = vgg19_pretrained.features


mapping = {
    'block1.0': 0,
    'block1.2': 2,
    'block2.0': 5,
    'block2.2': 7,
    'block3.0': 10,
    'block3.2': 12,
    'block4.0': 19,
    'block4.2': 21
}


for my_layer, vgg_idx in mapping.items():
    vgg_layer = vgg19_features[vgg_idx]
    if isinstance(vgg_layer, nn.Conv2d):
        weight = vgg_layer.weight.data
        bias = vgg_layer.bias.data
        # my_layer es un string tipo "blockX.Y", lo separamos para acceder a la secuencia
        block_name, layer_idx = my_layer.split('.')
        layer_idx = int(layer_idx)
        block = getattr(model, block_name)
        block[layer_idx].weight.data.copy_(weight)
        block[layer_idx].bias.data.copy_(bias)
        print(f"Pesos de {my_layer} asignados desde vgg19.features[{vgg_idx}]")


for name, param in model.named_parameters():
    if "classifier" not in name:
        param.requires_grad = False

print("\nModelo IntermediateCNN con pesos transferidos desde VGG19:")
print(model)
```

```python

model = ResIncepNet(num_classes=3)


resnet50 = models.resnet50(pretrained=True)
with torch.no_grad():
    model.conv1.weight.copy_(resnet50.conv1.weight)
    model.bn1.weight.copy_(resnet50.bn1.weight)
    model.bn1.bias.copy_(resnet50.bn1.bias)
print("ResNet50 stem weights loaded.")


# Load Inception v3 pre-trained weights (using the new weights API and aux_logits=True)
inceptionv3 = models.inception_v3(weights=models.Inception_V3_Weights.IMAGENET1K_V1, aux_logits=True)
mixed_5b = inceptionv3.Mixed_5b  # This is an InceptionA module

with torch.no_grad():
    # Branch 1: map branch1_conv to Mixed_5b.branch1x1.conv
    if model.branch1_conv.weight.shape == mixed_5b.branch1x1.conv.weight.shape:
        model.branch1_conv.weight.copy_(mixed_5b.branch1x1.conv.weight)
        model.branch1_conv.bias.copy_(mixed_5b.branch1x1.conv.bias)
        print("branch1_conv weights loaded from Mixed_5b.branch1x1")

    # Branch 2: map branch2_conv1 to branch5x5_1.conv and branch2_conv2 to branch5x5_2.conv
    if model.branch2_conv1.weight.shape == mixed_5b.branch5x5_1.conv.weight.shape:
        model.branch2_conv1.weight.copy_(mixed_5b.branch5x5_1.conv.weight)
        model.branch2_conv1.bias.copy_(mixed_5b.branch5x5_1.conv.bias)
    if model.branch2_conv2.weight.shape == mixed_5b.branch5x5_2.conv.weight.shape:
        model.branch2_conv2.weight.copy_(mixed_5b.branch5x5_2.conv.weight)
        model.branch2_conv2.bias.copy_(mixed_5b.branch5x5_2.conv.bias)
        print("branch2 weights loaded from Mixed_5b.branch5x5_1 and branch5x5_2")

    # Branch 3: map branch3_conv1, branch3_conv2, branch3_conv3 to branch3x3dbl_1, _2, _3 respectively
    if model.branch3_conv1.weight.shape == mixed_5b.branch3x3dbl_1.conv.weight.shape:
        model.branch3_conv1.weight.copy_(mixed_5b.branch3x3dbl_1.conv.weight)
        model.branch3_conv1.bias.copy_(mixed_5b.branch3x3dbl_1.conv.bias)
    if model.branch3_conv2.weight.shape == mixed_5b.branch3x3dbl_2.conv.weight.shape:
        model.branch3_conv2.weight.copy_(mixed_5b.branch3x3dbl_2.conv.weight)
        model.branch3_conv2.bias.copy_(mixed_5b.branch3x3dbl_2.conv.bias)
    if model.branch3_conv3.weight.shape == mixed_5b.branch3x3dbl_3.conv.weight.shape:
        model.branch3_conv3.weight.copy_(mixed_5b.branch3x3dbl_3.conv.weight)
        model.branch3_conv3.bias.copy_(mixed_5b.branch3x3dbl_3.conv.bias)
        print("branch3 weights loaded from Mixed_5b.branch3x3dbl")

    # Branch 4: map branch4_conv to branch_pool.conv
    if model.branch4_conv.weight.shape == mixed_5b.branch_pool.conv.weight.shape:
        model.branch4_conv.weight.copy_(mixed_5b.branch_pool.conv.weight)
        model.branch4_conv.bias.copy_(mixed_5b.branch_pool.conv.bias)
        print("branch4_conv weights loaded from Mixed_5b.branch_pool")

print("\nFinal model ready for fine-tuning:")
print(model)
```

```python
#@title Loss Function and Optimizer

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.00001)
```

```python
#@title Training

num_epochs = 3
for epoch in range(num_epochs):
    running_loss = 0.0
    for images, labels in trainloader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()

        # Forward + backward + optimize
        device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
        model.to(device)

        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {running_loss/len(trainloader)}")
```

```python
#@title Test the model

correct = 0
total = 0

with torch.no_grad():
    for images, labels in testloader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Accuracy of the model on the test images: {100 * correct / total}%")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset, Dataset
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns
import gc

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

#############################################
# Data Loading (loaded only once)
#############################################

# Utility: filter indices based on class limits
def filter_indices_with_limits(dataset, class_sample_limits):
    indices = []
    class_counts = {cls: 0 for cls in class_sample_limits.keys()}
    for i in range(len(dataset)):
        _, label = dataset[i]
        class_name = dataset.classes[label]
        if class_name in class_sample_limits:
            if class_counts[class_name] < class_sample_limits[class_name]:
                indices.append(i)
                class_counts[class_name] += 1
            if all(class_counts[c] >= limit for c, limit in class_sample_limits.items()):
                break
    return indices

# Define transforms (base transform converts PIL to tensor)
base_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5), (0.5,0.5,0.5))
])
# Augmentation for minority class
augmentation_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5), (0.5,0.5,0.5))
])

# Load CIFAR10 training data (without any transform initially)
trainset_full = datasets.CIFAR10(root='CIFAR10_data/', train=True, download=True, transform=None)

# Create an unbalanced training subset:
# 500 airplane, 500 automobile, 50 bird
train_limits = {'airplane': 500, 'automobile': 500, 'bird': 50}
train_indices = filter_indices_with_limits(trainset_full, train_limits)
trainset_unbalanced = Subset(trainset_full, train_indices)

# Create a balanced dataset by replicating minority ('bird') samples with augmentation
class BalancedAugmentedDataset(Dataset):
    def __init__(self, subset_dataset, base_transform, augmentation_transform, minority_class="bird", augmentation_factor=10):
        self.subset_dataset = subset_dataset
        self.base_transform = base_transform
        self.augmentation_transform = augmentation_transform
        self.minority_class = minority_class
        self.augmentation_factor = augmentation_factor
        self.indices = []
        for idx in range(len(self.subset_dataset)):
            image, label = self.subset_dataset[idx]
            class_name = self.subset_dataset.dataset.classes[label]
            if class_name == self.minority_class:
                self.indices.extend([(idx, True)] * self.augmentation_factor)
            else:
                self.indices.append((idx, False))
    def __len__(self):
        return len(self.indices)
    def __getitem__(self, index):
        idx, is_minority = self.indices[index]
        image, label = self.subset_dataset[idx]
        if is_minority:
            image = self.augmentation_transform(image)
        else:
            image = self.base_transform(image)
        return image, label

trainset_balanced = BalancedAugmentedDataset(trainset_unbalanced, base_transform, augmentation_transform,
                                              minority_class="bird", augmentation_factor=10)

# Create test set: filter CIFAR10 testset to 100 samples per class (airplane, automobile, bird)
testset_full = datasets.CIFAR10(root='CIFAR10_data/', train=False, download=True, transform=base_transform)
test_limits = {'airplane': 100, 'automobile': 100, 'bird': 100}
test_indices = filter_indices_with_limits(testset_full, test_limits)
testset_filtered = Subset(testset_full, test_indices)

# DataLoaders (using a modest batch size to help with memory)
batch_size = 16
trainloader_unbalanced = DataLoader(trainset_unbalanced, batch_size=batch_size, shuffle=True)
trainloader_balanced = DataLoader(trainset_balanced, batch_size=batch_size, shuffle=True)
testloader = DataLoader(testset_filtered, batch_size=batch_size, shuffle=False)
```

```python

def train_and_evaluate(model, train_loader, test_loader, num_epochs=3, lr=0.001):
    model.to(device)
    optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.9)
    criterion = nn.CrossEntropyLoss()
    model.train()
    for epoch in range(num_epochs):
        running_loss = 0.0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        print(f"Epoch {epoch+1}/{num_epochs} Loss: {running_loss/len(train_loader):.4f}")
        torch.cuda.empty_cache()
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for images, labels in test_loader:
            images = images.to(device)
            outputs = model(images)
            preds = outputs.argmax(dim=1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(labels.numpy())
    return np.array(all_labels), np.array(all_preds)

def plot_confusion(cm, title):
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

#############################################
# Experiment 1: IntermediateCNN with balanced dataset
#############################################
print("Experiment 1: IntermediateCNN with balanced dataset")
model_bal = IntermediateCNN(num_classes=3)
true_labels_bal, preds_bal = train_and_evaluate(model_bal, trainloader_balanced, testloader)
cm_bal = confusion_matrix(true_labels_bal, preds_bal)
plot_confusion(cm_bal, "IntermediateCNN - Balanced Data")

del model_bal
gc.collect()
torch.cuda.empty_cache()

```

```python


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def train_and_validate(model, train_loader, val_loader, num_epochs=5, lr=0.001):
    train_losses, val_losses = [], []
    model.to(device)
    optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.9)
    criterion = nn.CrossEntropyLoss()
    for epoch in range(num_epochs):
        model.train()
        running_train_loss = 0.0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_train_loss += loss.item()
        avg_train_loss = running_train_loss / len(train_loader)
        train_losses.append(avg_train_loss)

        model.eval()
        running_val_loss = 0.0
        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                running_val_loss += loss.item()
        avg_val_loss = running_val_loss / len(val_loader)
        val_losses.append(avg_val_loss)

        print(f"Epoch {epoch+1}/{num_epochs} - Train Loss: {avg_train_loss:.4f} - Val Loss: {avg_val_loss:.4f}")
        torch.cuda.empty_cache()
    return train_losses, val_losses

def plot_loss_curves(train_losses, val_losses, title="Training/Validation Loss"):
    epochs = range(1, len(train_losses)+1)
    plt.figure(figsize=(6,4))
    plt.plot(epochs, train_losses, 'b-o', label="Train Loss")
    plt.plot(epochs, val_losses, 'r-o', label="Validation Loss")
    plt.title(title)
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()

def evaluate_model(model, loader):
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for images, labels in loader:
            images = images.to(device)
            outputs = model(images)
            preds = outputs.argmax(dim=1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(labels.numpy())
    return np.array(all_labels), np.array(all_preds)

def compute_metrics(true_labels, preds):
    cm = confusion_matrix(true_labels, preds)
    num_classes = cm.shape[0]
    sensitivity_list, specificity_list = [], []
    for i in range(num_classes):
        TP = cm[i, i]
        FN = np.sum(cm[i, :]) - TP
        FP = np.sum(cm[:, i]) - TP
        TN = np.sum(cm) - (TP + FN + FP)
        sensitivity = TP / (TP + FN) if (TP+FN) != 0 else 0.0
        specificity = TN / (TN + FP) if (TN+FP) != 0 else 0.0
        sensitivity_list.append(sensitivity)
        specificity_list.append(specificity)
    # Macro-average across classes
    metrics = {
        "sensitivity": np.mean(sensitivity_list),
        "specificity": np.mean(specificity_list)
    }
    # Also compute F1 using sklearn's classification_report
    from sklearn.metrics import f1_score
    metrics["f1_score"] = f1_score(true_labels, preds, average="macro", zero_division=0)
    return cm, metrics

def plot_bar_metrics(metrics_dict, title="Evaluation Metrics"):
    labels = list(metrics_dict.keys())
    values = [metrics_dict[k] for k in labels]
    plt.figure(figsize=(5,4))
    plt.bar(labels, values, color="skyblue")
    plt.ylim([0, 1])
    plt.title(title)
    plt.ylabel("Score")
    plt.show()


#############################################
# 4. Train Models on Balanced and Unbalanced Datasets
#############################################
# Instantiate models:
netA_bal = IntermediateCNN(num_classes=3)
netA_unbal = IntermediateCNN(num_classes=3)
netB_bal = ResIncepNet(num_classes=3)
netB_unbal = ResIncepNet(num_classes=3)

print("Training IntermediateCNN on Balanced Data")
train_losses_A_bal, val_losses_A_bal = train_and_validate(netA_bal, trainloader_balanced, valloader_balanced, num_epochs=3)
plot_loss_curves(train_losses_A_bal, val_losses_A_bal, title="IntermediateCNN (Balanced) Loss")

print("Training IntermediateCNN on Unbalanced Data")
train_losses_A_unbal, val_losses_A_unbal = train_and_validate(netA_unbal, trainloader_unbalanced, valloader_unbalanced, num_epochs=3)
plot_loss_curves(train_losses_A_unbal, val_losses_A_unbal, title="IntermediateCNN (Unbalanced) Loss")

print("Training ResIncepNet on Balanced Data")
train_losses_B_bal, val_losses_B_bal = train_and_validate(netB_bal, trainloader_balanced, valloader_balanced, num_epochs=3)
plot_loss_curves(train_losses_B_bal, val_losses_B_bal, title="ResIncepNet (Balanced) Loss")

print("Training ResIncepNet on Unbalanced Data")
train_losses_B_unbal, val_losses_B_unbal = train_and_validate(netB_unbal, trainloader_unbalanced, valloader_unbalanced, num_epochs=3)
plot_loss_curves(train_losses_B_unbal, val_losses_B_unbal, title="ResIncepNet (Unbalanced) Loss")

#############################################
# 5. Evaluate and Compare Performance
#############################################
print("Evaluating IntermediateCNN (Balanced)")
true_A_bal, preds_A_bal = evaluate_model(netA_bal, testloader)
cm_A_bal, metrics_A_bal = compute_metrics(true_A_bal, preds_A_bal)
sns.heatmap(cm_A_bal, annot=True, fmt="d", cmap="Blues")
plt.title("IntermediateCNN (Balanced) Confusion Matrix")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.show()
plot_bar_metrics(metrics_A_bal, title="IntermediateCNN (Balanced) Metrics")

print("Evaluating IntermediateCNN (Unbalanced)")
true_A_unbal, preds_A_unbal = evaluate_model(netA_unbal, testloader)
cm_A_unbal, metrics_A_unbal = compute_metrics(true_A_unbal, preds_A_unbal)
sns.heatmap(cm_A_unbal, annot=True, fmt="d", cmap="Blues")
plt.title("IntermediateCNN (Unbalanced) Confusion Matrix")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.show()
plot_bar_metrics(metrics_A_unbal, title="IntermediateCNN (Unbalanced) Metrics")

print("Evaluating ResIncepNet (Balanced)")
true_B_bal, preds_B_bal = evaluate_model(netB_bal, testloader)
cm_B_bal, metrics_B_bal = compute_metrics(true_B_bal, preds_B_bal)
sns.heatmap(cm_B_bal, annot=True, fmt="d", cmap="Blues")
plt.title("ResIncepNet (Balanced) Confusion Matrix")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.show()
plot_bar_metrics(metrics_B_bal, title="ResIncepNet (Balanced) Metrics")

print("Evaluating ResIncepNet (Unbalanced)")
true_B_unbal, preds_B_unbal = evaluate_model(netB_unbal, testloader)
cm_B_unbal, metrics_B_unbal = compute_metrics(true_B_unbal, preds_B_unbal)
sns.heatmap(cm_B_unbal, annot=True, fmt="d", cmap="Blues")
plt.title("ResIncepNet (Unbalanced) Confusion Matrix")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.show()
plot_bar_metrics(metrics_B_unbal, title="ResIncepNet (Unbalanced) Metrics")


```
